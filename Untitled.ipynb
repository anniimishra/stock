```python
import pandas as pd
df=pd.read_csv('Data.csv', encoding='ISO-8859-1')
df.head()
```

```python
train=df[df['Date']<'20150101']
test=df[df['Date']>'20141231']
```

```python
data=train.iloc[:,2:27]
data.replace("[^a-zA-Z^]"," ",regex=True,inplace=True)
```

```python
list1=[i for i in range(25)]
new_index=[str(i) for i in list1]
data.columns=new_index
data.head()
```

```python
for index in new_index:
    data[index]=data[index].str.lower()
```

```python
headline=[]
for row in range(0,len(data.index)):
    headline.append(' '.join(str(x) for x in data.iloc[row,0:25]))
headline[4]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
countvector=CountVectorizer(ngram_range=(2,2))
traindataset=countvector.fit_transform(headline)
```

```python
random=RandomForestClassifier(n_estimators=200,criterion='entropy')
random.fit(traindataset,train['Label'])
```

```python
test_transform=[]
for row in range(0,len(test.index)):
    test_transform.append(" ".join(str(x) for x in test.iloc[row,2:27]))
test_data=countvector.fit_transform(test_transform)
```

```python
pred=random.predict(test_data)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
score=accuracy_score(test['Label'],pred)
score
matrix= confusion_matrix(test['Label'],pred)
matrix
```
